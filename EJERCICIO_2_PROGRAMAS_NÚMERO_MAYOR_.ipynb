```python
#Determinar el número mayor de entre 3 variables
num1 = int(input("Ingrese el primer número: "))
num2 = int(input("Ingrese el segundo número: "))
num3 = int(input("Ingrese el tercer número: "))

# Inicializar la variable 'mayor' con el primer número
mayor = num1

# Comparar num2 con el actual 'mayor'
if num2 > mayor:
    mayor = num2

# Comparar num3 con el actual 'mayor'
if num3 > mayor:
    mayor = num3

# Imprimir el resultado
print(f"\nEl número mayor de los tres es: {mayor}")
```

**Variables utilizadas**

1.   "num1": almacena el primer número ingresado, puede ser *int* o *float*.
2.   "num2": almacena el segundo número ingresado, puede ser *int* o *float*.
3.   "num3": almacena el tercer número ingresado, puede ser *int* o *float*.
4.   "mayor": almacena el número mayor de los tres ingresados anteriormente.

**Descripción en lenguaje natural**

Inicio del programa.

Pedir al usuario que ingrese el primer número y almacenarlo en num1.

Pedir al usuario que ingrese el segundo número y almacenarlo en num2.

Asumir inicialmente que num1 es el mayor.

Comparar si num2 es mayor que el mayor actual. Si lo es, num2 pasa a ser el nuevo mayor.

Comparar si num3 es mayor que el mayor actual. Si lo es, num3 pasa a ser el nuevo mayor.

Mostrar el valor almacenado en la variable mayor.

Fin del programa.
