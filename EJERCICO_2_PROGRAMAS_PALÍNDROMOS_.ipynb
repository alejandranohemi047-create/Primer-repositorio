```python
import re

#Definir si una frase o palabra es un palíndromo

def es_palindromo(palabra_original):
    # 1. Limpiar el texto: convertir a minúsculas y eliminar caracteres no alfanuméricos
    # Se usa re.sub para eliminar todo lo que no sea una letra o número
    texto_limpio = re.sub(r'[^a-z0-9]', '', palabra_original.lower())

    # 2. Invertir el texto limpio
    # [::-1] es una técnica de slicing en Python para invertir cadenas
    texto_invertido = texto_limpio[::-1]

    # 3. Comparar
    return texto_limpio == texto_invertido

# Solicitar la entrada al usuario
palabra = input("Ingrese una palabra o frase para verificar si es un palíndromo: ")

# Evaluar y mostrar el resultado
if es_palindromo(palabra):
    print(f"\n✅ '{palabra}' es un palíndromo.")
else:
    print(f"\n❌ '{palabra}' NO es un palíndromo.")
```

**Variables utilizadas**

1.   "palabra_original": Almacena la palabra o frase ingresada por el usuario.
2.   "texto_limpio": Almacena la palabra_original convertida a minúsculas y sin espacios (o signos de puntuación).
3.   "texto_invertido": Almacena la versión invertida de texto_limpio.
